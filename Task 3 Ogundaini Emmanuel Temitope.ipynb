## Handling Missing Values

```python
#calling out the order details id

data_set["order_details_id"].value_counts()
```

```python
#Changing to numerical values
data_set["order_details_id"] = pd.to_numeric(data_set["order_details_id"], errors = "coerce")
```

```python
#Filling the order details id with mean
data_set["order_details_id"] = data_set["order_details_id"].fillna(data_set["order_details_id"].mean())
```

```python
data_set.isna().sum()
```

**Note that our order_details_id is now free of nun values above.**

**We do same step for all other nun values so as to clean our dataset**

```python
#calling out the order id

data_set["order_id"].value_counts()
```

- **Filling the nun values for order_id and Quantity simultaneously**

```python
#Using method fill NA

data_set["order_id"] = pd.to_numeric(data_set["order_id"], errors = "coerce")
data_set["quantity"] = pd.to_numeric(data_set["quantity"], errors = "coerce")
```

```python
#Filling the order_id  and quantity with mean
data_set["order_id"] = data_set["order_id"].fillna(data_set["order_id"].mean())
data_set["quantity"] = data_set["quantity"].fillna(data_set["quantity"].mean())
```

```python
#Checking for nun values

data_set.isna().sum()
```

- **Filling the nun values for DATE & TIME**

```python
data_set["date"].value_counts()
```

```python
data_set ["time"].value_counts()
```

```python
#Filling the values with a fixed date and time 

data_set["date"] = data_set["date"].fillna(pd.to_datetime("2015-03-22"))

data_set["time"] = data_set["time"].fillna(pd.to_datetime("12:53:29"))
```

```python
#Checking if there are other nunvalues

data_set.isna().sum()
```

### 7.Find the total sales made on each day of the week,which day of the week is when sales are made the most?

```python
#first 3 rows of our data set

data_set.head(3)
```

```python
#EXTRACTING THE DAY OF THE WEEK FROM OUR DATE

data_set ["day_of_week"] = data_set["date"].dt.day_name()
```

```python
#CALCULATING THE TOTAL SALES PER DAY OF THE WEEK

sales_per_day = data_set.groupby("day_of_week")["Revenue"].sum().sort_values(ascending = False)

sales_per_day
```

- **Getting the day of the week with maxmimum sales**

```python
#DAY OF THE WEEK WHERE SALES ARE MADE THE MOST

best_sales_day = sales_per_day.idxmax()

print("Day with most sales:", best_sales_day)
```

**Day with most sales is Friday**

### VISUAL REPRESENTATION OF SALES_PER_DAY

```python
import matplotlib.pyplot as plt
```

```python
#PLOTING THE BARCHAT

plt.bar(sales_per_day.index,sales_per_day.values)
plt.xticks(rotation= 45)
plt.title("Total Sales by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Total Sales Revenue")

plt.show()
        
```

## 9. Find the sales made in each month. any trend noticeable?

- **Extract the months from the date column**

```python
#MONTHS IN WORDS

data_set["Month_in_words"] = data_set["date"].dt.month_name()
```

```python
#TOTAL SALES REVENUE PER MONTH

sales_per_month = data_set.groupby("Month_in_words")["Revenue"].sum()

#REORDER MONTHS IN CALENDAR ORDER

months_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
sales_per_month = sales_per_month.reindex(months_order)

sales_per_month
```

### VISUAL REPRESENTATION OF SALES BY MONTH

```python
import matplotlib.pyplot as plt
```

```python
#PLOTING THE BARCHAT

plt.bar(sales_per_month.index,sales_per_month.values)
plt.xticks(rotation= 45)
plt.title("Montly Sales Trend")
plt.xlabel("Months")
plt.ylabel("Total Sales Revenue")

plt.show()
```

**From the above visual representation of the monthly sales trend it is observed that:**

 - **There are seasonal spikes in the montly sales at intervals.**
 - **There was sales spike in January, March, May, July and November.**
 - **It was observed that September and October had almost equal number of sales.**
 - **The month of July had more sales.**
